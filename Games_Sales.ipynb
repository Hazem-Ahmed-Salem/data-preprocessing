```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
categorical_cols = ['Platform', 'Genre', 'Publisher','Game_Name']
for col in categorical_cols:
    df[col + '_encoded'] = label_encoder.fit_transform(df[col])


print("\nEncoded categorical variables:(HEAD)\n")
print(df[['Platform', 'Platform_encoded', 'Genre', 'Genre_encoded', 'Publisher', 'Publisher_encoded', 'Game_Name', 'Game_Name_encoded']].head(10))
print("\nEncoded categorical variables:(TAIL)\n")
print(df[['Platform', 'Platform_encoded', 'Genre', 'Genre_encoded', 'Publisher', 'Publisher_encoded']].tail(10))
```

```python
df.drop(["Game_Name","Platform","Genre","Publisher"],axis=1,inplace=True)
```

```python
q1 = df['NA_Sales'].quantile(0.25)
q1
```

```python
q3 = df['NA_Sales'].quantile(0.75)
q3
```

```python
iqr = q3 - q1
iqr
```

```python
lower = q1 - 1.5 * iqr
lower
```

```python
upper = q3 + 1.5 * iqr
upper
```

```python
df = df[(df['NA_Sales'] >= lower) & (df['NA_Sales'] <= upper)]
df
```
